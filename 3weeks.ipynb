```python
from thinkdsp import play_wave, read_wave
from IPython.display import Audio
import thinkplot
```

### 다운로드 한 Sound sample

```python
wave = read_wave('piano_sound.wav')
wave.make_audio()
```

```python
wave.plot()
```

### 0.5초 분량의 Segment

```python
start = 0
duration = 0.5
segment = wave.segment(start, duration)
segment.make_audio()
```

```python
segment.plot()
```

### 선택한 Segment의 spectrum

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

### low_pass 사용해보기

```python
spectrum.low_pass(5000)
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
```

```python
#정규화 작업
filtered.normalize()
```

#### 부드럽게 앞뒤도 깎아줌, apodize 사용

```python
filtered.apodize()
filtered.plot()
```

```python
filtered.make_audio()
```

### 오리지날 사운드랑 비교 해보기

```python
segment.make_audio()
```

### 비교결과 음이 오리지날 사운드에 비해, 낮아진것 같습니다.

### 이번엔 추가로, high_pass 적용해보기

```python
spectrum.high_pass(500)
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
filtered.plot()
```

```python
filtered.make_audio()
```

### 역시 오리지날 사운드와 비교

```python
segment.make_audio()
```

### 오리지날에 비해, 낮은음들이 안들리는게 확실하게 느껴집니다.

### band_stop 사용해보기

```python
spectrum.band_stop(2000,3000)
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
filtered.plot()
```

```python
filtered.make_audio()
```
